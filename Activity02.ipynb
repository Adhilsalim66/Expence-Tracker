```python
file = "expenses.txt"

def add_expense():
    try:
        cat = input("Category: ")
        amt = input("Amount: ")
        float(amt)
        with open(file, "a") as f:
            f.write(f"{cat},{amt}\n")
        print("Added!\n")
    except ValueError:
        print("Invalid amount!\n")

def view_expenses():
    try:
        with open(file, "r") as f:
            print("\nExpenses:")
            for line in f:
                cat, amt = line.strip().split(",")
                print(f"{cat} - {amt}")
        print()
    except FileNotFoundError:
        print("No file found!\n")

def calculate_total():
    try:
        with open(file, "r") as f:
            total = sum(float(line.strip().split(",")[1]) for line in f)
            print(f"Total = {total}\n")
    except (FileNotFoundError, ValueError):
        print("No records found!\n")

while True:
    print("1. Add\n2. View\n3. Total\n4. Exit")
    ch = input("Choice: ")
    if ch == "1":
        add_expense()
    elif ch == "2":
        view_expenses()
    elif ch == "3":
        calculate_total()
    elif ch == "4":
        break
    else:
        print("Wrong choice\n")
```
